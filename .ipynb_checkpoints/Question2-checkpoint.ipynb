```python
#load data

MR = pd.read_excel('UNIGME-2024-Country-Rates-Deaths-Under-five.xlsx',header=10)

MR.head()

```

```python
NR = pd.read_excel('UNIGME-2024-Country-Rates-Deaths-Under-five.xlsx', sheet_name=2, header=10)

NR.head()
```

```python
# Filter for EAC countries in NR dataframe (similar to what was done for MR)
eac_countries = ["Burundi", "Kenya", "Rwanda", "South Sudan", "Tanzania", "Uganda", "Democratic Republic of the Congo", "Somalia"]
eac_nr_data = NR[NR["Country.Name"].isin(eac_countries)]

# Identify all NMR columns (from 1950 to 2023) in NR dataframe
nmr_columns = [col for col in eac_nr_data.columns if col.startswith('NMR.')]

# Melt the NR DataFrame to transform year columns into rows
eac_nr_data_melted = eac_nr_data.melt(
    id_vars=['ISO.Code', 'Country.Name', 'SDG.Region', 'SDG.Subregion', 
             'UNICEF.Region', 'UNICEF.Programme.Region', 'Uncertainty.Bounds*'],
    value_vars=nmr_columns,
    var_name='year',
    value_name='NMR_value'
)

# Extract just the year part from the 'year' column
eac_nr_data_melted['year'] = eac_nr_data_melted['year'].str.replace('NMR.', '').astype(int)

# Sort the data for better readability
eac_nr_data_melted = eac_nr_data_melted.sort_values(['Country.Name', 'year'])
```

```python
# First, let's identify all U5MR columns (from 1950 to 2023)
u5mr_columns = [col for col in eac_data.columns if col.startswith('U5MR.')]

# Melt the DataFrame to transform year columns into rows
eac_data_melted = eac_data.melt(
    id_vars=['ISO.Code', 'Country.Name', 'SDG.Region', 'SDG.Subregion', 
             'UNICEF.Region', 'UNICEF.Programme.Region', 'Uncertainty.Bounds*'],
    value_vars=u5mr_columns,  # Use all identified U5MR columns
    var_name='year',
    value_name='U5MR_value'
)

# Extract just the year part from the 'year' column
eac_data_melted['year'] = eac_data_melted['year'].str.replace('U5MR.', '').astype(int)

# Sort the data for better readability
eac_data_melted = eac_data_melted.sort_values(['Country.Name', 'year'])

# Display the result
eac_data_melted
```

```python
eac_data_melted['Country.Name'].value_counts()
```

```python
import os

# List of ISO codes for the 8 EAC countries
eac_iso_codes = ['BDI', 'KEN', 'RWA', 'SSD', 'UGA', 'COD', 'SOM']
```

```python
# Path to the folder with the downloaded shapefiles
folder_path = "eac_shapefiles/"
```

```python
# Load and merge all shapefiles
gdfs = []
for iso in eac_iso_codes:
    file_path = os.path.join(folder_path, f"gadm41_{iso}_shp.zip")
    gdf = gpd.read_file(f"zip://{file_path}", layer=f"gadm41_{iso}_0")  # explicitly Level 0
    gdfs.append(gdf)
```

```python
# Combine all into one GeoDataFrame
eac_gdf = gpd.GeoDataFrame(pd.concat(gdfs, ignore_index=True))
```

```python
eac_gdf.plot()
```

```python
eac_gdf
```

```python
# Get latest year (assuming data is sorted)
latest_data = eac_data_melted[eac_data_melted["year"] == eac_data_melted["year"].max()]
latest_data
```

```python
latest_nr_data = eac_nr_data_melted[eac_nr_data_melted['year'] == 2023]
latest_nr_data
```

#####  Merge mortality data with shapefile

```python
# Merge the MR and NR data for 2023
merged_2023_data = pd.merge(
    latest_data,
    latest_nr_data,
    on=['ISO.Code', 'Country.Name', 'SDG.Region', 'SDG.Subregion', 
        'UNICEF.Region', 'UNICEF.Programme.Region', 'Uncertainty.Bounds*', 'year'],
    how='outer'
)
```

```python
# Merge with the shapefile
eac_gdf_with_data = eac_gdf.merge(
    merged_2023_data,
    left_on='GID_0',
    right_on='ISO.Code',
    how='left'
)


eac_gdf_with_data
```

###### Visualize the latest estimate of each indicator at the country level using shapefiles

```python
eac_map=eac_gdf_with_data.loc[eac_gdf_with_data['Uncertainty.Bounds*']== 'Median']

# 4. Plot choropleth maps
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

# Under-5 Mortality Rate (USMR)
eac_map.plot(column='U5MR_value', ax=ax1, legend=True,
         cmap='OrRd', legend_kwds={'label': 'USMR (Median)'})
ax1.set_title('Under-5 Mortality Rate by Country')

# Neonatal Mortality Rate (NMR)
eac_map.plot(column='NMR_value', ax=ax2, legend=True,
         cmap='Blues', legend_kwds={'label': 'NMR (Median)'})
ax2.set_title('Neonatal Mortality Rate by Country')

plt.tight_layout()
plt.show()
```

```python
merged_mr_nr = pd.merge(
    eac_nr_data_melted,
    eac_data_melted,
    on=['ISO.Code', 'Country.Name', 'SDG.Region', 'SDG.Subregion', 
        'UNICEF.Region', 'UNICEF.Programme.Region', 'Uncertainty.Bounds*', 'year'],
    how='outer'
)

merged_mr_nr
```

```python
# --- Plot U5MR ---
plt.figure(figsize=(10,6))

# Plot individual country points
sns.scatterplot(data=merged_mr_nr, x='year', y='U5MR_value', hue='Country.Name', legend=False, alpha=0.4)

# Plot the average trend
sns.lineplot(data=merged_mr_nr.groupby('year')['U5MR_value'].mean().reset_index(),
             x='year', y='U5MR_value', color='black', linewidth=2, label='Average U5MR')

# Finalize plot
plt.title('Average Trend of Under-5 Mortality Rate (U5MR) Over Time')
plt.ylabel('U5MR (per 1,000 live births)')
plt.xlabel('Year')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# --- Plot NMR ---
plt.figure(figsize=(10,6))

# Plot individual country points
sns.scatterplot(data=merged_mr_nr, x='year', y='NMR_value', hue='Country.Name', legend=False, alpha=0.4)

# Plot the average trend
sns.lineplot(data=merged_mr_nr.groupby('year')['NMR_value'].mean().reset_index(),
             x='year', y='NMR_value', color='black', linewidth=2, label='Average NMR')

# Finalize plot
plt.title('Average Trend of Neonatal Mortality Rate (NMR) Over Time')
plt.ylabel('NMR (per 1,000 live births)')
plt.xlabel('Year')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# --- Map for Under-5 Mortality Rate (U5MR) ---

# Filter only Under-5 mortality
u5mr_data = eac_gdf_with_data[eac_gdf_with_data['Indicator'] == 'Under-five mortality rate']

fig, ax = plt.subplots(1, 1, figsize=(10, 8))
u5mr_data.plot(column='Observation Value', cmap='OrRd', legend=True, ax=ax, edgecolor='black')
ax.set_title('Under-5 Mortality Rate (U5MR) - Latest Estimate')
ax.axis('off')
plt.show()

# --- Map for Neonatal Mortality Rate (NMR) ---

# Filter only Neonatal mortality
nmr_data = eac_gdf_with_data[eac_gdf_with_data['Indicator'] == 'Neonatal mortality rate']

fig, ax = plt.subplots(1, 1, figsize=(10, 8))
nmr_data.plot(column='Observation Value', cmap='PuBu', legend=True, ax=ax, edgecolor='black')
ax.set_title('Neonatal Mortality Rate (NMR) - Latest Estimate')
ax.axis('off')
plt.show()
```
